```python
auto_quant
```

```python
# (b) What is the range of each quantitative predictor?
# (c) What is the mean and standard deviation of each quantitative predictor?

pd.DataFrame({'range': auto_quant.max() - auto_quant.min(),
              'mean': auto_quant.mean(),
              'std': auto_quant.std()})
```

```python
# (d) Now remove the 10th through 85th observations. 
# What is the range, mean, and standard deviation of each predictor 
# in the subset of the data that remains?

# drop observations 10 to 85
auto_quant_d = auto_quant.drop(auto_quant.index[10:85])
pd.DataFrame({'range': auto_quant_d.max() - auto_quant_d.min(),
              'mean': auto_quant_d.mean(),
              'std': auto_quant_d.std()})
```

```python
#10 (a)
boston = pd.read_csv('Boston_corrected.csv',index_col=0)
```

```python
boston.head()
```

```python
boston.shape #506 rows and 20 columns
```

```python
import seaborn as sns
sns.pairplot(boston)
```

```python
#10 (d) Range of per capita crime rate by town

crime_range_max=boston.groupby('TOWN').CRIM.max()
crime_range_min=boston.groupby('TOWN').CRIM.min()
crime_range = crime_range_max - crime_range_min
crime_range.sort_values(ascending=False)
```

```python
#10 (d) Range of per capita crime rate by town
#Max value
boston.iloc[boston['CRIM'].argmax()]
```

```python
#Min value
boston.iloc[boston['CRIM'].argmin()]
```

```python
#10 (e) suburbs bound by the Charles river

print(boston[boston['CHAS']==1].TOWN.unique())
print('and their number: ',boston[boston['CHAS']==1].TOWN.nunique())
```

```python
#10 (f) medians pupil-teacher ratio among the towns

boston.PTRATIO.median()
```
